# Design

## Defining the business problem to be solved

Begginer speedcubers (competitive Rubik's cube solvers) can often struggle to understand how different factors - such as years of experience, number of competitions entered, geographical region, and participation frequency - influence their solving performance. While many enthusiest will already track their data using their own personal bests or averages, there is currently no widely accessible system that uses historical competition data and statistics to **predict a competitor's likely performance level**.

instead, cubers and coaches will typically rely on anecdotal experience or limited personal data when trying to estimate who much they can potentially improve. This makes it somewhat difficult for competitiors to:

*  set realistic performance goals
*  understand how their experience compares to other competitors
*  Identify whether competition exposure improves performance

This project aims to create a **data-driven prediction model** that estimates a competitiors average 3x3x3 competition solve time using historical competition statistics. By studying factors such as experience level, competition participation, participation frequency, and competition history, the model will ooutput an estimated performance level for the cuber. This tool could help begginer and intermediate cubers better understand their likely performance tragectory in in competitions, and make more informed decisions on training, competition participation, and competitor/s and their skill levels. 

## Refactoring the business problem into a machine learning problem


The business problem can be labeled as a **supervised regression machine learning problem**.

Given a cubers country, competitive longevity, participation frequency, completed events, and performance progression, predict a **continuous numerical value: the cubers average 3x3x3 solve time** - allows begginer-intermediate competitive speedcubers understand their likely performance in competitions, and make better decisions about training and improvement. 

This value represents the competitors overall solving ability and is widely used as a benchmark for performance in speed-cubing competitions. The machine learning model will therefore attempt to learn the relationship between the cuber's experience, perticipation statistics, adn their average solve time. 

## Defining success metrics

Success metrics for this project should focus souly on whether or not the model provides meaningful predictions that are practical and useful to the intended users rather than just being purely statistical measures. 

The model will be considered a general success if it achieves the following outcomes: 

*  **Realistic performance predictions**
    <br/>
    Predicted solve times should fall within a reasonable range of competition results. If the model consistenyl produces unrealistsiic predictions, for example extremley fast or extremely slow times, then it would have failed to provide meaningful insights.  
    <br/>
*  **Useful Accuracy for Begginer and Intermediate Cubers**
    <br/>
    The model should be accurate enough to provide practical guidance for begginer and intermediate competitors. Predictions should generally fall within a small margin of the actual solve time so that users can estimate their expectied performance levels.  
    <br/>

## Researching available data.

To build the prediction model, a dataset containing competition statistics was sourced from publicly available speedcubing competition records. Competitive speedcubing data is maintained by the **World Cube Association (WCA)**, which records official results from all competitions worldwide.

The dataset used for this project contains statistics about competitors, inluding their participation history and competitive activity. 


**Dataset Overview**

- The dataset contains the following fields: 

     | Column | Data |
     | ------ | ------ |
     | Average    | The competitors average 3x3x3 solve time for that particular competition |
     | Person ID    | A unique identifier asigned to each competitor |
     | Person name | The submitted name for the cmopetitor sompeting |
     | Gender | The gender of the competitor - M\F |
     | Person country ID | The country the competitior is from |
     | Best   | The competitors best 3x3x3 solve time for that particular competition |
     | competition ID   | The ID of the competition participated in |
     | Event ID | The name of the competition participated in |
     | Year | The year the competition was held in |

These fetures can be wrangled and engineered to capture important aspect's of a competitors experience level and engagement in the competition community.

**Relevence of the Data** 

The dataset is suitable for this project because it provides many measurable indicators of competition activity that may influence solving performance

For example:
* Competitors who attend **more competitions** may gain more experience under pressure or in the timed, comp environment
* Cubers who have been **active for more years** indicates a strong development in more complex solving techniques 
* A **higher number of events** completed may indicate a **broader experience** with official competition environments, even if the event isn't 3x3x3
* Evaluating **competitions per year** reflects how frequently a competitor actively practices in competition settings.  

By analysing these features together, a machine learning model can learn patterns that can realte experience and participation to a cubers solving ability. 

**Data Limitations**

Although the dataset contains many useful indicators of competitive experience, there are several limitations that may effect prediction accuracy: 

* The dataset does not contain practice time or training habbits leading up to the competition, which does stringly influence cubing performance
* The dataset does not contain age information, which in some cases may correlate to learning speed. (However, some domain knowledge requires you to know that the current world record holder is a 9 year old polish boy, so age does not really have a strong effect on this model)
* Differences in hardware/cube type are not recorded (the better cube model, the more features it has, the more lubricant it has, are all domain knowledge features that would effect a cubers performance)

Despite these limitations, the dataset still provides valuable and structured data that can reveal **general performance trends within the speedcubing commmunity**.
