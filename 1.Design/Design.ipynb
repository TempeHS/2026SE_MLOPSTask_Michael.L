# Design

## Defining the business problem to be solved

Begginer speedcubers (competitive Rubik's cube solvers) can often struggle to understand how different factors - such as years active, demographic, or age - influence their solve times. Enthusiests will currently rely on previous and anecdotal data rather than structured predictions, which makes it hard to plan training or track expected progress. I want to provide a data-driven prediction tool that predicts a competitor's 3x3x3 solve time based on historical competition data. This would allow coaches and competitors to understand performance trends, identify possible areas for improvement, or make informed decisions about implementable training strategies. 

## Refactoring the business problem into a machine learning problem
*First describe the problem in plain language, write down the problem in simple, direct words requires thinking through what really needs to be accomplished. It is important you identify what business outcome you are trying to achieve, not what process or model you want to use.*

Given a cubers country, competitive longevity, participation frequency, completed events, and performance progression, predict their average 3x3x3 solve time to allow begginer-intermediate competitive speedcubers understand their likely performance, and make better decisions about training and improvement. 

## Defining success metrics

*We are not talking about MSE, MAE, or R2 score. The success metrics should be focused on business, addressing the ideal outcome you identified earlier. How will you determine if the machine learning model fails to live up to business requirements? How will you know it has succeeded?*

> The information and ranges provided below are to help students understand the domain of the data, it is not intended as medical or diagnostic advice.
